## Filing Type

```python
master_index_all_states_df['Filing Type'].nunique()
```

```python
# keywords for rate
# rate
# loss cost
# LCM
# rating
# model? # need to manually check
# check legal issue
```

```python
rate_related_keywords = ['rate','loss cost','lcm','rating','model']
```

```python
filtered_df = master_index_all_states_df[master_index_all_states_df['Filing Type'].apply(
    lambda x: any(keyword in x.lower() for keyword in rate_related_keywords)
)]

# Output the shape of the filtered DataFrame
print(filtered_df.shape)
```

```python
round(filtered_df.shape[0]/master_index_all_states_df.shape[0],2)
```

```python
filtered_df.head()
```

```python
count_by_state = filtered_df.groupby('State')['Filing Type'].count()

plt.figure(figsize=(16, 8))

count_by_state.plot(kind='bar', color='deepskyblue')

plt.xlabel('State')
plt.ylabel('Number of Filings')
plt.title('Number of Filings by State')
plt.xticks(rotation=45)
plt.show()
```
